```python
# отбрасываем не заполненные колонки
drop_columns = ["Alley", "MiscVal", "MiscFeature", "Fence", "PoolQC", "FireplaceQu", "MasVnrType"]
df1 = df.drop(labels=drop_columns, axis='columns')
df1.info()
```

```python
# заполняем отсутствующие  значения колонок NA
def na_mode_replace(df: pd.DataFrame, col_name_list: list):
    for col_name in col_name_list:
        print("--------------")
        print(f"Column '{col_name}', before correction:")
        for value in df1["GarageCond"].unique():
            print(value, ":", df1[df1["GarageCond"]==value].GarageCond.count())
        mode_garage_conditions = df[col_name].mode()[0]
        df[col_name] = df[col_name].fillna("NA")
        print(f"Column '{col_name}', after correction:")
        print(df[col_name].describe())
        print("--------------")
    return df
df2 = df1.copy()
cols_list = ["GarageType", "GarageFinish"]
df2 = na_mode_replace(df2, col_name_list=cols_list)
    
```

```python
# заполняем отсутствующие  значения колонок модой
def na_mode_replace(df: pd.DataFrame, col_name_list: list):
    for col_name in col_name_list:
        print("--------------")
        print(f"Column '{col_name}', before correction:")
        for value in df1["GarageCond"].unique():
            print(value, ":", df1[df1["GarageCond"]==value].GarageCond.count())
        mode_garage_conditions = df[col_name].mode()[0]
        df[col_name] = df[col_name].fillna(mode_garage_conditions)
        print(f"Column '{col_name}', after correction:")
        print(df[col_name].describe())
        print("--------------")
    return df
cols_list = ["GarageCond", "GarageQual", "BsmtQual", "BsmtCond", "BsmtExposure", "BsmtFinType1", "BsmtFinType2", "Electrical"]
df2 = na_mode_replace(df2, col_name_list=cols_list)
    
```

```python
# заполняем отсутствующие  значения колонок медианой
def na_median_replace(df: pd.DataFrame, col_name_list: list):
    for col_name in col_name_list:
        median_garage_conditions = df[col_name].median()
        df[col_name] = df[col_name].fillna(median_garage_conditions)
        print(df[col_name].describe())
    return df
cols_list = ["LotFrontage", "GarageYrBlt", "MasVnrArea"]
df2 = na_median_replace(df2, col_name_list=cols_list)
```

```python
# преобразуем дуальные типы  в числовые
convert_dict = {"CentralAir": {"Y":1, "N":0}, "Utilities": {"AllPub": 1, "NoSeWa":0}, "Street": {"Pave":1, "Grvl":0 }}
cols_list = ["CentralAir", "Utilities", "Street"]
def create_num_cols(df, cols_list):
    work_df = df.copy()
    for col in cols_list:
        replace_dict = convert_dict.get(col)
        work_df[f"{col}_num"] = work_df[col].map(replace_dict)
    return work_df
```

```python
# удалим выбросы в колонках
from scipy import stats
import numpy as np
cols_list = ["SalePrice", "OpenPorchSF", "LotFrontage", "LotArea", "1stFlrSF", "WoodDeckSF", "MasVnrArea", "BsmtFinSF1", "GarageArea"]
def delete_outlaiers(df, cols_list):
    work_df = df.copy()
    for col in cols_list:
        z = np.abs(stats.zscore(work_df[col]))
        threshold = 3
        outliers = work_df[col][z > threshold]
        if not outliers.empty:
            print(f"Column {col} has {len(outliers)} outlaiesr")
        work_df = work_df.drop(outliers.index)
        # work_df.loc[z > threshold, col] = work_df[col].median()
    return work_df
```
